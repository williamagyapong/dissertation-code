### Data Cleaning

```python
def clean_df(df, redundant_col_number=31):
    # Strangely this "Unamed {redundant_col_number}" column was appended to the data
    df = df.copy()
    df = df[df['HINCP'] > 0] # limit data to $1 or more income

    income_bins = [0, 25000, 50000, 75000, 100000, 200000, float('inf')]
    income_labels = [1, 2, 3, 4, 5, 6]
    df['HINCP'] = pd.cut(df['HINCP'], bins=income_bins, labels=income_labels, right=False)
    df['AGEP'] = pd.cut(df['AGEP'],
                        bins=[0,5, 18, 25, 35, 45, 65, 85, float('inf')],
                        labels=[1, 2, 3, 4, 5, 6, 7, 8], right=False
                        ) 
    
    # Recode variables with less represented levels
    df['NP'] = df['NP'].apply(lambda x: 9 if x >= 9 else x)
    df['JWRIP'] = df['JWRIP'].apply(lambda x: 4 if x >= 4 else x)
    # Drop redundant columns
    # 'WIF', 
    df.drop(columns=['WGTP', 'PWGTP', 'ST', 'VACS', f'Unnamed: {redundant_col_number}'], inplace=True) 

    return df
```

```python
texas.columns
```

```python
df = clean_df(texas, redundant_col_number=30) # 30 is the index of the redundant column
print(texas.shape)
print(df.shape)
df.head()
```

```python
# Save treated data
df.to_csv('Texas/texas_treated.csv', index=None)
```
